```python
%matplotlib notebook
import matplotlib.pyplot as plt

import numpy as np

#bc no relative import from __main__
# TODO open jupyter with -m flag?
import sys as sys
sys.path.append('..')

from bayes.linearparametric import BayesLinear as Model
from bayes.linearparametric.bases import gaussianType as phi
```

### The Gaussian Basis Function

```python
phiV = 1.
xGB = np.arange(90.)/30. - 1.5
plt.plot(xGB, phi(np.expand_dims(xGB, axis=1), 0, phiV))
plt.ylabel('Gaussian Type Basis')
```

## The Test Data

```python
noiseStdDev = 0.2
x = np.arange(100.)/10.
t = np.sin(x/1.5) + np.random.normal(0, noiseStdDev, size=x.size)
plt.scatter(x, t)
plt.ylabel('Gaussian Type Basis')
```

## The Fit!

```python
fit = Model(phi, [{'u':np.array([i]), 'v':np.array(phiV)} for i in range(10)], noiseStdDev=noiseStdDev)
fit.train(np.expand_dims(x, axis=1),t)
```

```python
plt.plot(x, fit.predictive(np.expand_dims(x, axis=1))[0])
#plt.scatter(x, fit.sample(np.expand_dims(x, axis=1)))
```
