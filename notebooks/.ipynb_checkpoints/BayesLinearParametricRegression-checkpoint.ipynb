### The Gaussian Basis Function

```python
phiV = 1.
xGB = np.arange(90.)/30. - 1.5
plt.plot(xGB, phi(xGB, 0, phiV))
plt.ylabel('Gaussian Type Basis')
```

## The Test Data

```python
noiseStdDev = 0.2
x = np.arange(100.)/10.
t = np.sin(x/1.5) + np.random.normal(0, noiseStdDev, size=x.size)
plt.scatter(x, t)
plt.ylabel('Gaussian Type Basis')
```
